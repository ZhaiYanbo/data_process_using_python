### 数据描述：
数据集中的一些字段描述：

	Rank- 按收入中位数排名（数据集按此列排序）。

	Major_code-专业代码。

	Major- 专业名称。

	Major_category- 专业类别。

	Total- 主修该专业的总人数。

	Sample_size-计算收入相关的数据时采样的样本总量（无加权）。

	Men-男毕业生总数

	Women-女毕业生总数

	ShareWomen- 女性毕业生占比。

	Employed- 就业人数。

	Median- 该专业全年全职的工作的毕业生的薪资中位数。

	Low_wage_jobs- 从事低薪服务性工作的人数

	Full_time-工作时间为35小时以上人数。

	Part_time-工作时间少于35小时人数。


## 第一关：数据清洗

```python
import pandas as pd
import matplotlib.pyplot as plt
#以便内联显示绘图
%matplotlib inline 
recent_grads = pd.read_csv('recent-grads.csv')
print(recent_grads.iloc[0])
print(recent_grads.head())
print(recent_grads.tail())
recent_grads.describe()
```
